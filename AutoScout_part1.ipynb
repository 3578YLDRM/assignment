```python
columns_to_drop = ['MakeModel','ShortDescription','OfferNumber','Colour','EmptyWeight','UpholsteryColour','OtherFuelTypes','EmissionsSticker','ProductionDate',\
                   'ManufacturerColour','FuelConsumptionWltp','CoEmissionsWltp','AvailableFrom','TaxiOrRentalCar','Availability','LastTimingBeltChange',\
                   'ElectricRangeWltp','PowerConsumptionWltp','BatteryOwnership','EmissionClass','PreviousOwner'        
]

df = df.drop(columns=columns_to_drop)
```

```python
df
```

```python
df['Make']= df['Make'].str.strip()
df['Price'] = df['Price'].str.replace('[^\d.]', '', regex=True).astype(float)
df['Mileage']=df['Mileage'].str.replace(',','').str.extract('(\d{1,8})')
df['Warranty']=df['Warranty'].str.extract('(\d{1,3})')
df['CoEmissions']=df['CoEmissions'].str.extract('(\d{1,3})')
df
```

```python
df.dropna(axis=0, how="all", inplace=True)
```

```python
df
```

```python
def clean3(value):
    if isinstance(value, list) and len(value) > 0:
        first_element = value[0]
        if isinstance(first_element, str):
            return first_element[3:-3]
    return value
```

```python
def clean1(value):
    if isinstance(value, list) and len(value) > 0:
        first_element = value[0]
        if isinstance(first_element, str):
            return first_element[1:-1]
    return value
```

```python
def clean0(value):
    if isinstance(value, list) and len(value) > 0:
        first_element = value[0]
        if isinstance(first_element, str):
            return first_element[1:-3]
    return value
```

```python
columns_to_clean = ['Model','BodyType','Type','Doors','CountryVersion','Warranty','Seats','Drivetrain','ModelCode']

df[columns_to_clean] = df[columns_to_clean].applymap(clean3)
```

```python
columns_to_clean = ['Gearbox','Power','Gears','Cylinders']

df[columns_to_clean] = df[columns_to_clean].applymap(clean1)
```

```python
columns_to_clean = ['EngineSize']

df[columns_to_clean] = df[columns_to_clean].applymap(clean0)
```
